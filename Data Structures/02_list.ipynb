Mutable – Mutability refers to whether an object's value can be changed after creation. And List allows this.

Duplicates – We know data structures are used to store multiple values, so duplicates means the same value occurring multiple times. List allows this.

Ordered – List maintains the sequence of elements as they were inserted. This means you can access elements using their position (index).

Heterogeneous – Lists have a heterogeneous nature, which means we can have multiple data types inside the list.

Traversing over list

```python
lst = [1,2,3,4,5]

for i in range(len(lst)):
    print(lst[i])
```

Methods of list

```python
numbers = [5, 4, 1, 5, 6]   # Initial list
print("Initial List:", numbers)

# append()
numbers.append(10)
print("After append(10):", numbers)

# insert()
numbers.insert(2, 15)
print("After insert(2, 15):", numbers)

# extend()
numbers.extend([20, 25, 30])
print("After extend([20, 25, 30]):", numbers)

# remove()
numbers.remove(5)
print("After remove(5):", numbers)

# pop()
popped_item = numbers.pop()
print("Popped Item:", popped_item)
print("After pop():", numbers)

# index()
index = numbers.index(6)
print("Index of 6:", index)

# count()
count_5 = numbers.count(5)
print("Count of 5:", count_5)

# sort()
numbers.sort()
print("After sort():", numbers)

# reverse()
numbers.reverse()
print("After reverse():", numbers)

# copy()
new_numbers = numbers.copy()
print("Copied List:", new_numbers)

# clear()
numbers.clear()
print("After clear():", numbers)
```
