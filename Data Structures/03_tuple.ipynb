Tuples

Immutable - Tuples are not mutable you cannot change the values of tuple.

Duplicates - You can have duplicate values in tuple there are no restriction.

Ordered - Set are ordered and you can access them through index values.

Heterogenous - Set also have heterogenous nature and can have different types of data structure in tuple.

```python
a = (1,2,3,4)

print(type(a))
```

Traversal is same as of list traversal.

Python tuples have only 2 built-in methods because tuples are immutable

```python
t = (10, 20, 10, 30, 10)

print(t.count(10))   # 3
print(t.count(20))   # 1
print(t.count(40))   # 0
```

```python
t = (10, 20, 30, 20, 40)

print(t.index(20))      # 1
print(t.index(30))      # 2
print(t.index(20, 2))   # 3
```

Tuple unpacking (or multiple assignment).

```python
a, b, c, d = (1, 2, 3, 4)

print(a)
print(b)
print(c)
print(d)
```
