Sets

Mutable – Sets are mutable, so you can add or remove elements.

Duplicates – Sets do not allow duplicate elements. Every element is unique.

Unordered – Sets are unordered, so they do not support indexing or slicing. So it cant be traversed also.

Heterogeneous – Sets can store different immutable data types together (e.g., integers, floats, strings, tuples, booleans). However, mutable objects like lists, dictionaries, and other sets cannot be elements of a set because they are unhashable.

```python
a = {1,2,3,4,4,5,5,6,5}

print(type(a))
print(a)  #As it stores only unique value so duplicate vlues will be removed.
```

```python
a = {1,2,3,4,4,5,5,6,5}

print(a[0])  # We cant access set value using index 
```
