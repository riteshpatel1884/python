```python
def hello(first_name, last_name):
    print(first_name, last_name)

hello("Ritesh", "Patel")
```

```python
def hello(first_name, age):
    print(first_name, age)

hello(21, "Ritesh")
```

```python
# To avoid it we use keyword arguement
def hello(first_name, age):
    print(first_name, age)

hello(age = 21, first_name = "Ritesh")

```

```python
def hello(first_name, last_name, age = 20):
    print(first_name, last_name, age)

hello(first_name= "Ritesh", last_name = "Patel")
```

```python
def hello(first_name, last_name, age = 20):
    print(first_name, last_name, age)

hello(first_name= "Ritesh", last_name = "Patel", age = 21)
```

```python
def hello():
    return "hi"

print(hello())
```
