Range
Range(start, stop, step)
start = 0 = default value
step = 1 = default value

For loop

```python
for i in range(1,20,3):
    print(i)
```

```python
a = range(1,20,4)
for i in a:
    print(i)
```

```python
a = range(10)  # start = 0 and step = 1
for i in a:
    print(i)
```

```python
a = range(10,0,-1) 
for i in a:
    print(i)
```

```python
a = range(-5,-15,-1) 
for i in a:
    print(i)
```

```python
a = "Ritesh Patel"
for i in range(len(a)):  # length = 12
    print(a[i])
```

```python
a = "Ritesh Patel"
for i in a:  
    print(i) # now we are iterating ovre characters not index
```

While Loop

```python
a = 1
while(a<=10):
    print(a)
    a = a+1
```
